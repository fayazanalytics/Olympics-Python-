```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Olympics.csv")
```

```python
df_gym = df[df["Sport"] == 'Gymnastics']
df_lifting = df[df["Sport"] == 'Weightlifting']
```

```python
#Create a scatter plot between age and weight based on gender.
```

```python
Gender = df['Sex']
```

```python
plt.figure(figsize = (8,6))
sns.set_style("whitegrid")
x = sns.boxplot(x = df['Weight'], palette = 'Set1')
x.set(xlabel = 'Weight', title = 'Boxplot for Weight')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.set_style("whitegrid")
x = sns.boxplot(x = df['Age'], palette = 'Set1')
x.set(xlabel = 'Age', title = 'Boxplot for Age')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.set_style("whitegrid")
x = sns.boxplot(x = df['Height'], palette = 'Set1')
x.set(xlabel = 'Height', title = 'Boxplot for Height')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.set_style('whitegrid')
x = sns.scatterplot(x = df['Age'], y = df['Weight'], hue = df['Sex'])
x.set(xlabel = 'Age', title = 'Scattering Weight and Age on Gender')
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.set_style("whitegrid")
x = sns.scatterplot(x = df['Height'], y = df['Weight'],data = df_gym,hue = df["Sex"],palette = 'Set1')
x.set(title= "Scatter plot for Height and Weight based on Gender for Gymnastics")
plt.show()
```

```python
sns.set_style = ("dark", {'axes.grid': False})
plt.figure(figsize = (8,6))
x = sns.barplot(x = df_top3['Sex'], y = df_top3['Medal'],hue = df_top3['Sex'], estimator = sum, errorbar = None)
x.set(xticklabels = ['Gender'], title = "Total Medals")
plt.xlabel = 'Gender'
plt.ylabel = "Medals"
plt.tight_layout
plt.show()
```
